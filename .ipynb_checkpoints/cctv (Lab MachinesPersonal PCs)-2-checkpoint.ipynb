Create a function that allows us to play video frames in Colab envrinment. 

```python


def display_frames(frames):
  fig=plt.figure(figsize=(10,6))
  mov=[]
  for i in range(len(frames)):
    img=plt.imshow(frames[i],animated=True)
    plt.axis('off')
    mov.append([img])

  anime = animation.ArtistAnimation(fig,mov,interval=50,repeat_delay=1000)

  plt.close()
  return anime
```

Extract all video frames

```python
vidPath=r'video.mp4'
cap=cv2.VideoCapture(vidPath)
frames=[]
i=0
while(1):
  ret,frame=cap.read()
  if ret:
    frame=cv2.resize(frame,dsize=(600,400))
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    frames.append(frame)
  else:
    break
cap.release()
```

Play video frames as HTML video (this is only necessary for Colab envrionment. For standard python development use OpenCV's playback fundtion)

```python
HTML(display_frames(frames).to_html5_video())
```

Take the first two frames and plot their differences as binary image.

```python
frame0_float=np.float32(frames[0])
frame1_float=np.float32(frames[1])

frame_diff=(frame0_float-frame1_float)**2
frame_diff_allchannel=np.sum(frame_diff,axis=2)
```

A function that allows us to play frame-difference frame by frame with an additional threshold to tune results.

```python
def display_diff_frames(frames,threshold):
  fig=plt.figure(figsize=(10,6))
  mov=[]
  for i in range(len(frames)):
    img=plt.imshow(frames[i]>int(threshold),cmap='gray',animated=True)
    plt.axis('off')
    mov.append([img])

  anime = animation.ArtistAnimation(fig,mov,interval=50,repeat_delay=1000)

  plt.close()
  return anime
```

Going through all video frames and calculate differences
* only save results when sufficient differences are detected
*   use morphological filtering to remove noise
*   use cvlib to detect object



```python
import cvlib
from cvlib.object_detection import draw_bbox
n=[]
diff_frames=[]
for i, frame in enumerate(frames):

  if i==0:
    left=np.float32(frame)
  else:
    diff_frame=(np.float32(frame)-left)**2
    diff_allchannel=np.sum(diff_frame,axis=2)
    k_r=cv2.getStructuringElement(cv2.MORPH_ERODE,(3,3))
    diff_allchannel=cv2.morphologyEx(np.float32(diff_allchannel),cv2.MORPH_ERODE,k_r)
    if diff_allchannel.mean()>10:
      diff_frames.append(diff_allchannel)
      bbox,labels,conf=cvlib.detect_common_objects(frame)
      frame=draw_bbox(frame,bbox,labels,conf)
      n.append(frame)
    left=np.float32(frame)
    

#HTML(display_diff_frames(diff_frames,1500).to_html5_video())
```
